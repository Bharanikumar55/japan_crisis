```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# File paths
pop_path = "../data/raw/japan_total_population.csv"
elderly_path = "../data/raw/japan_elderly_percent.csv"
gdp_path = "../data/raw/japan_gdp.csv"
fertility_path = "../data/raw/japan_fertility_rate.csv"

# Load CSVs (skip first 4 metadata rows)
pop = pd.read_csv(pop_path, skiprows=4)
elderly = pd.read_csv(elderly_path, skiprows=4)
gdp = pd.read_csv(gdp_path, skiprows=4)
fertility = pd.read_csv(fertility_path, skiprows=4)

pop.head()
```

```python
pop = pop[pop["Country Name"] == "Japan"]
elderly = elderly[elderly["Country Name"] == "Japan"]
gdp = gdp[gdp["Country Name"] == "Japan"]
fertility = fertility[fertility["Country Name"] == "Japan"]
```

```python
cols_to_drop = ["Country Name", "Country Code", 
                "Indicator Name", "Indicator Code"]

pop = pop.drop(columns=cols_to_drop)
elderly = elderly.drop(columns=cols_to_drop)
gdp = gdp.drop(columns=cols_to_drop)
fertility = fertility.drop(columns=cols_to_drop)
```

```python
pop = pop.T
elderly = elderly.T
gdp = gdp.T
fertility = fertility.T
```

```python
# Convert Year column safely

pop = pop[pd.to_numeric(pop["Year"], errors="coerce").notnull()]
elderly = elderly[pd.to_numeric(elderly["Year"], errors="coerce").notnull()]
gdp = gdp[pd.to_numeric(gdp["Year"], errors="coerce").notnull()]
fertility = fertility[pd.to_numeric(fertility["Year"], errors="coerce").notnull()]

# Now convert to integer
pop["Year"] = pop["Year"].astype(int)
elderly["Year"] = elderly["Year"].astype(int)
gdp["Year"] = gdp["Year"].astype(int)
fertility["Year"] = fertility["Year"].astype(int)
```

```python
pop = pop.reset_index()
elderly = elderly.reset_index()
gdp = gdp.reset_index()
fertility = fertility.reset_index()

pop.columns = ["Year", "Total_Population"]
elderly.columns = ["Year", "Elderly_Percent"]
gdp.columns = ["Year", "GDP"]
fertility.columns = ["Year", "Fertility_Rate"]
```

```python
pop = pop[(pop["Year"] >= 1990)]
elderly = elderly[(elderly["Year"] >= 1990)]
gdp = gdp[(gdp["Year"] >= 1990)]
fertility = fertility[(fertility["Year"] >= 1990)]
```

```python
df = pop.merge(elderly, on="Year")
df = df.merge(gdp, on="Year")
df = df.merge(fertility, on="Year")

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method="ffill")
```

```python
df.to_csv("../data/processed/japan_cleaned_data.csv", index=False)
```
