```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_csv("../data/processed/japan_cleaned_data.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Total_Population"])
plt.title("Japan Total Population Trend")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```

### Insight 1:
Japan’s population peaked around 2010 and has been declining since then.
This confirms the demographic contraction trend.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Elderly_Percent"])
plt.title("Elderly Population Percentage (65+)")
plt.xlabel("Year")
plt.ylabel("Elderly %")
plt.show()
```

### Insight 2:
The percentage of population aged 65+ has been increasing continuously.
Japan is transitioning into a super-aged society.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Fertility_Rate"])
plt.title("Fertility Rate Trend")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.show()
```

### Insight 3:
Japan’s fertility rate has remained significantly below replacement level (2.1),
indicating long-term population decline.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["GDP"])
plt.title("GDP Trend")
plt.xlabel("Year")
plt.ylabel("GDP (Current USD)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## 📌 Key Observations from EDA

1. Japan's population peaked around 2010 and is declining.
2. Elderly population percentage is steadily increasing.
3. Fertility rate remains below replacement level.
4. GDP growth appears stagnant in recent years.
5. Strong correlations exist between demographic factors.

These trends confirm the severity of Japan’s demographic crisis.
